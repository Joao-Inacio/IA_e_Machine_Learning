# **EXERCÍCIO**

- O objetivo desta tarefa é utilizar o haarcascade fullbody.xml e a imagem pessoas.jpg para detectar o corpo inteiro das pessoas



## Bibliotecas

```python
# Libs
import cv2
import matplotlib.pyplot as plt
```

```python
# Carregando a imagem
# Imagem BGR
img = cv2.imread('../data/pessoas.jpg')
# Imagem RGB
imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
# Visualizando a imagem
plt.figure(figsize=(10,10))
plt.title('Imagem Original')
plt.imshow(imgRGB);
```

```python
# Importando o modelo
detector_corpos= cv2.CascadeClassifier('../data/fullbody.xml')
```

```python
# Convertendo a imagem para escala de cinza
imagem_cinza = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_cinzaRGB = cv2.cvtColor(imagem_cinza, cv2.COLOR_GRAY2RGB)
plt.figure(figsize=(10,10))
plt.title('Imagem Cinza')
plt.imshow(img_cinzaRGB)
plt.show()
```

```python
# Variáveis de detecção 
deteccoes =detector_corpos.detectMultiScale(imagem_cinza, scaleFactor=1.01, minSize=(60, 50))
# Tamanho da matriz 
len(deteccoes)
```

```python
for (x, y, l, a) in deteccoes:
    #print(x, y, l, a)
    cv2.rectangle(imgRGB, (x, y), (x + l, y + a), (0,255,0), 2)
plt.figure(figsize=(10,10))
plt.title('Imagem Cinza')
plt.imshow(imgRGB)
plt.show()
```
